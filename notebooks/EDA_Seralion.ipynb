```python
# Loop through all numeric columns and detect outliers using IQR
numeric_columns = df.select_dtypes(include=['float64', 'int64']).columns

outliers_summary = {}

for column in numeric_columns:
    Q1 = df[column].quantile(0.25)  # 25th percentile
    Q3 = df[column].quantile(0.75)  # 75th percentile
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    # Count outliers
    outliers_count = df[(df[column] < lower_bound) | (df[column] > upper_bound)].shape[0]
    outliers_summary[column] = outliers_count

# Display the number of outliers per column
print("Outlier counts for each column:\n", outliers_summary)
plt.figure(figsize=(12, 6))
plt.bar(outliers_summary.keys(), outliers_summary.values(), color='skyblue')
plt.xticks(rotation=45, ha='right')
plt.title("Outlier Counts for Each Column", fontsize=16)
plt.xlabel("Columns", fontsize=14)
plt.ylabel("Number of Outliers", fontsize=14)
plt.tight_layout()
plt.show()
```

```python
# Create boxplots for all numeric columns
plt.figure(figsize=(15, 10))
sns.boxplot(data=df[numeric_columns])
plt.xticks(rotation=45)
plt.title("Boxplots for All Numeric Columns")
plt.show()
```

```python
# Define validation rules for each column
validation_rules = {
    'GHI': lambda x: x >= 0,
    'DNI': lambda x: x >= 0,
    'DHI': lambda x: x >= 0,
    'ModA': lambda x: x >= 0,
    'ModB': lambda x: x >= 0,
    'Tamb': lambda x: -50 <= x <= 50,
    'RH': lambda x: 0 <= x <= 100,
    'WS': lambda x: x >= 0,
    'WSgust': lambda x: x >= 0,
    'WSstdev': lambda x: x >= 0,
    'WD': lambda x: 0 <= x <= 360,
    'WDstdev': lambda x: x >= 0,
    'BP': lambda x: 900 <= x <= 1100,
    'Cleaning': lambda x: x in [0, 1],
    'Precipitation': lambda x: x >= 0,
    'TModA': lambda x: -50 <= x <= 50,
    'TModB': lambda x: -50 <= x <= 50,
}

# Identify incorrect entries for each column
incorrect_values = {}

for column, rule in validation_rules.items():
    if column in df.columns:  # Ensure column exists in the dataset
        incorrect_values[column] = df[~df[column].apply(rule)]  # Rows where the rule is violated
# Display incorrect values for each column
for column, incorrect_rows in incorrect_values.items():
    print(f"Incorrect values in {column}:\n", incorrect_rows)

incorrect_counts = {}

for column, rule in validation_rules.items():
    if column in df.columns:  # Ensure the column exists in the dataset
        incorrect_counts[column] = (~df[column].apply(rule)).sum()  # Count rows where the rule is violated

# Display the number of incorrect values for each column
incorrect_counts = {column: int(count) for column, count in incorrect_counts.items()}

print("Number of incorrect values in each column:\n", incorrect_counts)
```

```python
# Display the number of Incorrect Values per column
print("Outlier counts for each column:\n", incorrect_counts)
plt.figure(figsize=(12, 6))
plt.bar(incorrect_counts.keys(), incorrect_counts.values(), color='skyblue')
plt.xticks(rotation=45, ha='right')
plt.title("Incorrect Value Counts for Each Column", fontsize=16)
plt.xlabel("Columns", fontsize=14)
plt.ylabel("Number of Inncorect data", fontsize=14)
plt.tight_layout()
plt.show()
```
